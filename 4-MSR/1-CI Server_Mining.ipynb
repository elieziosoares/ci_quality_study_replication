```python
def searchFirstcommit(project):
    path = getTravisFilePath(project)
    if path is None:
        return None
    
    url = 'https://api.github.com/repos/{}/commits?path={}&per_page=100&page='.format(project,path)
    first_commit = None
    
    page = 1
    commits = requestAPI(url+str(page))
    
    
    while commits is not None:
        try:
            first_commit = commits[-1]['commit']['author']['date']
        
        except Exception as e:
            print('\n\n==============================================================')
            print('Error while processing storePR >>> Exception: {}'.format(e)) 
            print('Project: {}    PR - {}\n\n'.format(project, url+str(page)))
        
        page += 1
        commits= requestAPI(url+str(page))
    
    return first_commit
```

```python
def getTravisFilePath(project):
    gh='https://api.github.com/search/code?q=filename:.travis.yml+repo:'+project
    path = None
    
    try:
        r = requestAPI(gh)
    except Exception as e:
        print('{} -- requestAPI - {}'.format(datetime.now().strftime("%H:%M:%S"),e.message))
    
    if r is not None and 'total_count' in r and r['total_count'] > 0:
        return r['items'][0]['path']

    return None
```

```python
def searchCIService(project):
    if useGHWorkflows(project):
        return 'GH Workflows'
    elif useTravis(project):
        return 'Travis CI'
    elif useAppVeyor(project):
        return 'AppVeyor'
    elif useJenkins(project):
        return 'Jenkins'
    elif useWercker(project):
        return 'Wercker'
    elif useCircle(project):
        return 'Circle CI'
    else:
        return None
```

```python
def useTravis(project):
    gh='https://api.github.com/search/code?q=filename:.travis.yml+repo:'+project
    travis='https://api.travis-ci.org/repos/'+project
    
    try:
        r = requestTravisAPI(travis)
    except Exception as e:
        print('{} -- requestTravisAPI()'.format(datetime.now().strftime("%H:%M:%S")))
        r = None
            
    if r is not None and 'id' in r:
        return True
    else:
        try:
            r = requestAPI(gh)
        except Exception as e:
            print('{} -- requestAPI - {}'.format(datetime.now().strftime("%H:%M:%S"),e.message))
        
        if 'total_count' in r and r['total_count'] > 0:
            return True

    return False
```

```python
def useAppVeyor(project):
    gh='https://api.github.com/search/code?q=filename:appveyor.yml+repo:'+project

    r = requestAPI(gh)
    if 'total_count' in r and r['total_count'] > 0:
        return True

    return False
```

```python
def useJenkins(project):
    gh='https://api.github.com/search/code?q=filename:Jenkinsfile+repo:'+project

    r = requestAPI(gh)
    if 'total_count' in r and r['total_count'] > 0:
        return True

    return False
```

```python
def useWercker(project):
    wercker='https://app.wercker.com/api/v3/applications/' + project

    r = requestAPI(wercker)
    if 'id' in r:
        return True

    return False
```

```python
def useCircle(project):
    gh='https://api.github.com/search/code?q=path:.circleci+filename:config.yml+repo:'+project
    circle='https://circleci.com/api/v2/project/gh/' + project

    r = requestAPI(circle)
    if 'slug' in r:
        return True
    else:
        r = requestAPI(gh)
        if 'total_count' in r and r['total_count'] > 0:
            return True

    return False
```

```python
def useGHWorkflows(project):
    gh='https://api.github.com/search/code?q=path:.github/workflows+extension:.yml+repo:'+project

    r = requestAPI(gh)
    if 'total_count' in r and r['total_count'] > 0:
        return True

    return False
```

```python
import psycopg2
import os
import requests 
import time
from datetime import datetime, timedelta
import pandas as pd
from scipy import stats
import numpy as np
np.seterr(divide='ignore', invalid='ignore')
```

```python
def connectDB():
    f = open('/home/elieziosoares/psql_pwd.txt', "r")
    pwd = f.readline().replace('\n','')
    
    return psycopg2.connect(user = "postgres",
                              password = pwd,
                              host = "127.0.0.1",
                              port = "5432",
                              database = "Causal_CI_Quality_v3")
```

```python
def getProjects():
    query = """SELECT repo_name, analysis_point, created, last_commit From PROJECTS where ci_service is Null ORDER BY repo_name"""

    connection = connectDB()
    cursor = connection.cursor()
    cursor.execute(query)
    result = cursor.fetchall()
    connection.close()
    return result
```

```python
def getTravisProjects():
    query = """select repo_name from projects
            where qtd_stars >= 100 and forked is false and size  > 10000
            and ci_service like 'Travis CI' 
            and analysis_point is null"""
    
    connection = connectDB()
    cursor = connection.cursor()
    cursor.execute(query)
    result = cursor.fetchall()
    connection.close()
    return result

```

```python
def setProjectCIService(repo_name,service):
    connection = connectDB()
    cursor = connection.cursor()
    
    if service =='Travis CI':
        query = """UPDATE  projects 
        set ci_service = %s,
        travis_url = %s
        WHERE repo_name like %s"""
        
        cursor.execute(query, [service,'https://api.travis-ci.org/repos/'+repo_name,repo_name])
    else:
        query = """UPDATE  projects 
        set ci_service = %s
        WHERE repo_name like %s"""   
        cursor.execute(query, [service,repo_name])
        
    connection.commit()
    cursor.close()
    connection.close()
```

```python
def setCIProjectAnalysis_Point(repo_name,date):
    connection = connectDB()
    cursor = connection.cursor()
    
    query = """UPDATE  projects 
    set analysis_point = %s,
    travis_url = %s
    WHERE repo_name like %s"""

    cursor.execute(query, [date,'https://api.travis-ci.org/repos/'+repo_name,repo_name])
        
    connection.commit()
    cursor.close()
    connection.close()
```
