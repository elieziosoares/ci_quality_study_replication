```python
repos = getTravisProjects()
i=0
for repo in repos:    
    project = repo[0]
    
    repo_creation = getProjectCreation(project)[0][0]
    if repo_creation is None:
        repo_creation = getProjectCreationAPI(project)
        setRepoCreation(project,repo_creation)
    
    if repo_creation is None:
        continue
        
    travis_commit = getAnalysisPoint(project)[0][0]
    days = abs(travis_commit - repo_creation).days
    
    setProjectAnalysis_Point_Days(project,days)
        
    i += 1
    print('{}/{} - {} Project {} - Days: {}'.format(i,len(repos),datetime.now().strftime("%H:%M:%S"),project,days))
    
    
```

```python
repos = getProjectsMined()
i=0

for repo in repos:    
    project = repo[0]
    first_commit = repo[1]
    
    #Calculate init of the moth and finish of the last month
    #Adjust the finish date to the first hour of the subsequent month
    analysisInit = analysisFinish = datetime(first_commit.year, first_commit.month, 1, 0,0,0,tzinfo=first_commit.tzinfo)
    analysisFinish = analysisFinish + relativedelta(months=+13)
    
    
    setAnalysis_init_finish(project,analysisInit,analysisFinish)
        
    
    i += 1
    print('{}/{} - Project {} - Init: {} - Finish: {}'.format(i,len(repos),project,analysisInit,analysisFinish))
    #time.sleep(.02)
        
```

```python
def connectDB():
    f = open('/home/psql_pwd.txt', "r")
    pwd = f.readline().replace('\n','')
    
    return psycopg2.connect(user = "ci_quality",
                              password = pwd,
                              host = "127.0.0.1",
                              port = "5432",
                              database = "Causal_CI_Quality_v4")
```

```python
def getProjects():
    query = """SELECT repo_name, analysis_point, created, last_commit From PROJECTS 
    where ci_service is Null ORDER BY repo_name"""

    connection = connectDB()
    cursor = connection.cursor()
    cursor.execute(query)
    result = cursor.fetchall()
    connection.close()
    return result
```

```python
def getProjectsMined():
    query = """SELECT repo_name, analysis_point From PROJECTS 
    where  qtd_stars >= 100 and forked is false and size  > 10000
            and ((ci_service is null or ci_service like 'Travis CI') and analysis_point is not null)
            and releases_mined is true
            order by repo_name desc;"""

    connection = connectDB()
    cursor = connection.cursor()
    cursor.execute(query)
    result = cursor.fetchall()
    connection.close()
    return result
```

```python
def getNonCIProjects():
    query = """select repo_name, created from projects
            where qtd_stars >= 100 and forked is false and size  > 10000
            and ci_service is null 
            order by repo_name;"""
    
    connection = connectDB()
    cursor = connection.cursor()
    cursor.execute(query)
    result = cursor.fetchall()
    connection.close()
    return result
```

```python
def searchFirstcommit(project):
    path = getTravisFilePath(project)
    if path is None:
        return None
    
    url = 'https://api.github.com/repos/{}/commits?path={}&per_page=100&page='.format(project,path)
    first_commit = None
    
    page = 1
    commits = requestAPI(url+str(page))
    
    
    while commits is not None:
        try:
            first_commit = commits[-1]['commit']['author']['date']
        
        except Exception as e:
            print('\n\n==============================================================')
            print('Error while processing storePR >>> Exception: {}'.format(e)) 
            print('Project: {}    PR - {}\n\n'.format(project, url+str(page)))
        
        page += 1
        commits= requestAPI(url+str(page))
    
    return first_commit
```

```python
def getTravisFilePath(project):
    gh='https://api.github.com/search/code?q=filename:.travis.yml+repo:'+project
    path = None
    
    try:
        r = requestAPI(gh)
    except Exception as e:
        print('{} -- requestAPI - {}'.format(datetime.now().strftime("%H:%M:%S"),e.message))
    
    if r is not None and 'total_count' in r and r['total_count'] > 0:
        return r['items'][0]['path']

    return None
```

```python
def setProjectCIService(repo_name,service):
    connection = connectDB()
    cursor = connection.cursor()
    
    if service =='Travis CI':
        query = """UPDATE  projects 
        set ci_service = %s,
        travis_url = %s
        WHERE repo_name like %s"""
        
        cursor.execute(query, [service,'https://api.travis-ci.org/repos/'+repo_name,repo_name])
    else:
        query = """UPDATE  projects 
        set ci_service = %s
        WHERE repo_name like %s"""   
        cursor.execute(query, [service,repo_name])
        
    connection.commit()
    cursor.close()
    connection.close()
```

```python
def loadTokens():
    f = open('/home/gh_tokens.txt', "r")
    tokens =[]
    tk = f.readline().replace('\n','')
    while tk != '':
        tokens.append(tk)
        tk = f.readline().replace('\n','')

    f.close()
    return tokens
```

```python
tokens = loadTokens()
i_token =0


def requestAPI(URL):
    # api-endpoint 
    global i_token
    
    if (len(tokens)-1) > i_token:
        i_token += 1
    else:
        i_token = 0
        
    r = requests.get(url = URL, headers={'Authorization': tokens[i_token],'Accept':'application/vnd.github.cloak-preview'}).json()
    if len(r) > 0:
        try:
            #print('keys: {}'.format(r.keys()))
            
            if isinstance(r, dict) and 'message' in r.keys():
                print('{} -- {}'.format(datetime.now().strftime("%H:%M:%S"),r['message']))
                if 'API rate limit exceeded' in r['message']:
                    time.sleep(600)
                    requestAPI(URL)
                elif 'You have exceeded a secondary rate limit. Please wait a few minutes before you try again.' in r['message']:
                    time.sleep(10)
                    requestAPI(URL)
                else:
                    return None

            return r
        except Exception as e:
            print('\n Erro no request get: {}'.format(e))
            print(r)
    else:
        return None
```

```python
def requestTravisAPI(URL):
    
    r = requests.get(url = URL).json()
    return r
    
```

```python
def useGHWorkflows(project):
    gh='https://api.github.com/search/code?q=path:.github/workflows+extension:.yml+repo:'+project

    r = requestAPI(gh)
    if 'total_count' in r and r['total_count'] > 0:
        return True

    return False
```

```python
def useCircle(project):
    gh='https://api.github.com/search/code?q=path:.circleci+filename:config.yml+repo:'+project
    circle='https://circleci.com/api/v2/project/gh/' + project

    r = requestAPI(circle)
    if 'slug' in r:
        return True
    else:
        r = requestAPI(gh)
        if 'total_count' in r and r['total_count'] > 0:
            return True

    return False
```

```python
def useWercker(project):
    wercker='https://app.wercker.com/api/v3/applications/' + project

    r = requestAPI(wercker)
    if 'id' in r:
        return True

    return False
```

```python
def useJenkins(project):
    gh='https://api.github.com/search/code?q=filename:Jenkinsfile+repo:'+project

    r = requestAPI(gh)
    if 'total_count' in r and r['total_count'] > 0:
        return True

    return False
```

```python
def useAppVeyor(project):
    gh='https://api.github.com/search/code?q=filename:appveyor.yml+repo:'+project

    r = requestAPI(gh)
    if 'total_count' in r and r['total_count'] > 0:
        return True

    return False
```

```python
def useTravis(project):
    gh='https://api.github.com/search/code?q=filename:.travis.yml+repo:'+project
    travis='https://api.travis-ci.org/repos/'+project
    
    try:
        r = requestTravisAPI(travis)
    except Exception as e:
        print('{} -- requestTravisAPI()'.format(datetime.now().strftime("%H:%M:%S")))
        r = None
            
    if r is not None and 'id' in r:
        #print('\t\t\t ------> useTravis -- r is not None')
        return True
    else:
        try:
            r = requestAPI(gh)
        except Exception as e:
            print('{} -- requestAPI - {}'.format(datetime.now().strftime("%H:%M:%S"),e.message))
        
        if r is not None and 'total_count' in r and r['total_count'] > 0:
            #print('\t\t\t ------> useTravis -- total_count = {}'.format(r['total_count']))
            return True

    return False
```

```python
def searchCIService(project):
    if useGHWorkflows(project):
        return 'GH Workflows'
    if useTravis(project):
        return 'Travis CI'
    elif useAppVeyor(project):
        return 'AppVeyor'
    elif useJenkins(project):
        return 'Jenkins'
    elif useWercker(project):
        return 'Wercker'
    elif useCircle(project):
        return 'Circle CI'
    else:
        return None
```
