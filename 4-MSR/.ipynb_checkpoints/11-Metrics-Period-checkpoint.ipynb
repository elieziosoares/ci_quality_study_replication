```python
data = getProjectPeriods()
project = repo_creation = None

for project_period in data:  
    
    if project != project_period[0]:
        repo_creation = getProjectCreation(project_period[0])[0][0]
        #print(repo_creation)
        
    project = project_period[0]
    init = project_period[1]
    end = project_period[2]
    
    print('Project: {}\t\t{} - {}'.format(project,init,end))
    
    bugs = getBugs(project,init,end)
    bugs = int(bugs['bugs'][0])
    issues = getIssues(project,init,end)
    issues = int(issues['issues'][0])
    
    if issues == 0:
        bugs_ratio = 0
    else:
        bugs_ratio = bugs/issues
    
    print('\t\t bugs_ratio: {}'.format(bugs_ratio))
    
    
    ###################
    ## Bugs after tag and before a tag...
    ###################
    bugs = getBugsRQ2(project,init,end)
    qty_bugs = len(bugs)
    if qty_bugs > 0:
        try:
            qty_bugs_p1 = int(bugs['period'].value_counts()[1]) #counts values 1
        except Exception as e:
            qty_bugs_p1 = 0
        try:
            qty_bugs_p2 = int(bugs['period'].value_counts()[2]) #counts values 2
        except Exception as e:
            qty_bugs_p2 = 0
    else:
        qty_bugs_p1 = qty_bugs_p2 = 0
    
       
    age = abs(init - repo_creation).days
    prs = getPrs(project,init,end)
    communication = getCommunication(project,init,end)
    conflicts = getMergeConflicts(project,init,end)
    
    
    builds = getBuilds(project, init,end)
    commits = sum(map(lambda x: int(x), prs['commits'].tolist())) 
    
    #test_proportion ou test_volume
    try:
        test_vol_proportional = (prs['test_volume'].sum()/prs['commit_size'].sum())
    except Exception as e:
        test_vol_proportional = 0
    
    #print('{} - {} - {} - {}'.format(prs,communication,conflicts,bugs))
    #print('{}'.format(prs))
    #print('{}'.format(communication))
    #print('{}'.format(conflicts['conflicts'][0]))
    #print('{}'.format(bugs))
    #print(commits)

    updateMetricsPeriod(project,init,end,len(prs),prs['commit_size'].mean(),prs['commit_size'].median(),prs['commit_size_per_file'].mean(),prs['commit_size_per_file'].median(),(prs['test_volume']/prs['commit_size']).mean(),(prs['test_volume']/prs['commit_size']).median(),communication['communication'].mean(),communication['communication'].median(),int(conflicts['conflicts'][0]),bugs, len(builds),commits,test_vol_proportional,age,bugs_ratio,qty_bugs_p1,qty_bugs_p2)    
```

```python
def getProjects():    
    query = """SELECT repo_name, analysis_init, analysis_finish, analysis_point, rq2, rq2_analysis_init, rq2_analysis_finish, ci, best_analysis_finish From PROJECTS 
    WHERE rq1 is true;"""
    

    connection = connectDB()
    cursor = connection.cursor()
    cursor.execute(query)
    result = cursor.fetchall()
    connection.close()
    return result
```

```python
def getPrs(project_name,init,finish):
    connection = connectDB()
    query = 'select commit_size,(commit_size/files) commit_size_per_file,(test_volume/commit_size) TEST_PROPORTION, test_volume,commits from pullrequests WHERE (closed_at between %s and %s) AND project_name ilike %s AND commit_size is not null AND commit_size >0;'

    df = pd.read_sql_query(query,con=connection,params=[init,finish,project_name])
    connection.close()
    
    #Check if column contains all values 0. In this case we do not filter outliers.
    if checkValues(df['commit_size']) and checkValues(df['test_volume']):
        return df
    else:
        return removeOutliers(df)
    #return df
```

```python
def getCommunication(project_name,init,finish):
    connection = connectDB()
    query = 'select ("Comments"+review_comments) COMMUNICATION from pullrequests WHERE (closed_at between %s and %s) AND project_name ilike %s AND ("Comments" is not null or review_comments is not null);'

    df = pd.read_sql_query(query,con=connection,params=[init,finish,project_name])
    connection.close()
    
    #Check if column contains all values 0. In this case we do not filter outliers.
    if checkValues(df['communication']):
        return df
    else:
        return removeOutliers(df)
```

```python
def getMergeConflicts(project_name,initDate, finishDate):
    connection = connectDB()
    query = """select count(id) conflicts from PULLREQUESTS 
    where project_name ilike %s AND closed_at BETWEEN %s AND %s AND mergeconflict is True"""

    df = pd.read_sql_query(query,con=connection,params=[project_name,initDate, finishDate])
    connection.close()
    
    return df
```

```python
def getBugs(project_name,initDate, finishDate):
    query = """select count(id) bugs from ISSUE 
    where repo_name like %s AND created_at BETWEEN %s AND %s AND ISBUG IS TRUE;"""

    connection = connectDB()
    df = pd.read_sql_query(query,con=connection,params=[project_name,initDate, finishDate])
    connection.close()
    return df
```

```python
def getIssues(project_name,initDate, finishDate):
    query = """select count(id) issues from ISSUE 
    where repo_name like %s AND created_at BETWEEN %s AND %s;"""

    connection = connectDB()
    df = pd.read_sql_query(query,con=connection,params=[project_name,initDate, finishDate])
    connection.close()
    return df
```

```python
#https://www.kite.com/python/answers/how-to-remove-outliers-from-a-pandas-dataframe-in-python
def removeOutliers(df):
    z_scores = stats.zscore(df)
    abs_z_scores = np.abs(z_scores)
    filtered_entries = (abs_z_scores < 3).all(axis=1)
    df_filtered = df[filtered_entries]

    return df_filtered
```

```python
def checkValues(df):
    if (df == 0).all():
        return True
    else:
        i=v=0
        v = df[0]
        for a in df:
            if a != v:
                return True
            
        return False
```

```python
def updateMetricsPeriod(repo_name,initDate, finishDate,qtdPR,commit_size_mean,commit_size_median,commit_size_perfile_mean,commit_size_perfile_median,test_volume_mean,test_volume_median,communication_mean,communication_median,merge_conflicts,bugs,qtybuilds,commits,test_vol_proportional,age,bugs_ratio,qty_bugs_p1,qty_bugs_p2):
    try:
        query = """UPDATE  METRICS_PERIOD 
            set commit_size_mean = %s,
            commit_size_median = %s,
            commit_size_perfile_mean = %s,
            commit_size_perfile_median = %s,
            test_volume_mean = %s,
            test_volume_median = %s,
            communication_mean = %s,
            communication_median = %s,
            merge_conflicts = %s,
            bugs = %s,
            qtd_pull_request = %s,
            qty_builds =%s,
            qty_commits=%s,
            test_volume_proportional=%s,
            age=%s,
            bugs_ratio=%s,
            bugs_p1=%s,
            bugs_p2=%s
            WHERE repo_name like %s AND init_period = %s AND end_period = %s;"""

        connection = connectDB()
        cursor = connection.cursor()
        cursor.execute(query, [commit_size_mean,commit_size_median,commit_size_perfile_mean,commit_size_perfile_median,test_volume_mean,test_volume_median,communication_mean,communication_median,merge_conflicts,bugs,qtdPR,qtybuilds,commits,test_vol_proportional,age,bugs_ratio,qty_bugs_p1,qty_bugs_p2,repo_name,initDate, finishDate])
        connection.commit()
        connection.close()
    except psycopg2.IntegrityError as e:
        print ("==============================================================")
        print ("Error while updating into PostgreSQL. updateMetricsPeriod >>> Exception: {}".format(e)) 
        print('Project: {}    INIT - {}  - END - {}'.format(repo_name,initDate, finishDate))
        connection.close()
    except Exception as e:
        print ("==============================================================")
        print ("Error while processing updateMetricsPeriod >>> Exception: {}".format(e)) 
        print('Project: {}    INIT - {}  - END - {}'.format(repo_name,initDate, finishDate))
```

```python
def updateMetricsPeriodTMP(repo_name,initDate,finishDate,prs):
    try:
        query = """UPDATE  METRICS_PERIOD 
            set qtd_pull_request=%s
            WHERE repo_name like %s AND init_period = %s AND end_period = %s;"""

        connection = connectDB()
        cursor = connection.cursor()
        cursor.execute(query, [prs,repo_name,initDate, finishDate])
        connection.commit()
        connection.close()
    except psycopg2.IntegrityError as e:
        print ("==============================================================")
        print ("Error while updating into PostgreSQL. updateMetricsPeriod >>> Exception: {}".format(e)) 
        print('Project: {}    INIT - {}  - END - {}'.format(repo_name,initDate, finishDate))
        connection.close()
    except Exception as e:
        print ("==============================================================")
        print ("Error while processing updateMetricsPeriod >>> Exception: {}".format(e)) 
        print('Project: {}    INIT - {}  - END - {}'.format(repo_name,initDate, finishDate))
```

```python
def getBuilds(repo_name,initDate, finishDate):
    #print('****** Computing Build Health. Repo {}'.format(repo_name))
    
    query = """SELECT state FROM builds_mined WHERE 
            repo_name like %s AND FINISHED_at BETWEEN %s AND %s"""

    connection = connectDB()
    cursor = connection.cursor()
    
    cursor.execute(query, [repo_name,initDate, finishDate])
    rows = cursor.fetchall()
    
    return rows
```

```python
def getBugsRQ2(project,init,end):
    connection = connectDB()
    query = """SELECT period,days_after_tag,days_before_tag,days_between_tags,size,haslink,hasimage,comments FROM ISSUE 
    where repo_name like %s and isbug is true 
    and size is not null and days_between_tags > 0
    and created_at between %s and %s
    order by created_at;"""
    
    df = pd.read_sql_query(query,con=connection,params=[project,init,end])
    connection.close()
    
    return df
```
