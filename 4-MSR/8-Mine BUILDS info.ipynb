## Build Info Mining

1. Get preliminar info: Travis_repo_id, last_build_number, last_build_date
2. Get builds info

### 1. Request Travis info and get <travis_repo_id, last_build_number,last_build_date>

```python
projects = getProjects()
i=0
error=[]
while i < len(projects):
    try:
        #If we already have the information, skip to next one.
        if len(projects[i]) > 1:
            i+=1
            continue

        repo = projects[i][0]
        print('Getting info from {} ({}/{}) - {}.'.format(repo,i+1,len(projects),datetime.now().strftime("%D/%M/%Y %H:%M:%S")))

        # api-endpoint 
        URL = 'https://api.travis-ci.org/repos/{}'.format(repo)
        data = requestTravisAPI(URL)
        
        projects[i] = projects[i] + (data['id'], data['last_build_started_at'],data['last_build_number'])
        
        updateProjectRepo(projects[i][0],projects[i][1],projects[i][2],projects[i][3])
        print('Updated table projects info from {}: Repo id: {}; Builds: {}'.format(projects[i][0],projects[i][1],projects[i][2]))

        i+=1
    except KeyError:
        error.append(repo)
        print('ERROR WHILE REQUEST INFO FROM {} ({}/{}).'.format(repo,i+1,len(projects)))
        i+=1
#projects
```

### 2. Request Travis info of builds

```python
projects = getProjectsToMineBuilds()
i = 1
for p in projects:
    print('Getting info from {} ({}/{}) - {}.'.format(p[0],i,len(projects),datetime.now().strftime("%D/%M/%Y %H:%M:%S")))
    i+=1
    
    mineBuilds(p[0],p[1],int(p[2]))
```

## 3- Bind Builds to releases

```python
repos = getProjectsToMineBuilds()
i=0
for repo in repos:
    project = repo[0]
    i += 1
    print('\n\n\n{}/{} Project - {} - {}\n\n\n'.format(i,len(repos),project,datetime.now().strftime("%D/%M/%Y %H:%M:%S")))

    releases = getReleases(repo[0])
    
    j = len(releases) -1
    while j >= 0:
        release_creation = releases[j][2]
        release_id = releases[j][3]
        
        updateBuilds(project,release_creation,release_id)        
        j -= 1
```

```python
def getReleases(repo_name):
    connection = connectDB()
    cursor = connection.cursor()
    
    query = """select id,tag_name, created_at, node_id from project_releases
            where repo_name like %s
            ORDER BY created_at asc;"""
    
    cursor.execute(query, [repo_name])
    result = cursor.fetchall()
    connection.close()
    return result
```

```python
def updateBuilds(repo_name,dt,node_id):
    query = """select * from bindbuildsreleases(%s,%s,%s);"""
    
    #print('select * from bindbuildsreleases({},{},{});'.format(repo_name, dt,node_id))

    connection = connectDB()
    cursor = connection.cursor()
    cursor.execute(query, [repo_name, dt,node_id])
    connection.commit()
    cursor.close()
    connection.close()
```

```python
def mineBuilds(slug,repo_id,build_numb):
    #print('****** Starting mine repo {} from build number: {}'.format(repo_id,build_numb))
    
    #Query the database to continue mining in break situations
    build_min = getBuildNumber(repo_id)
    if build_min is not None:
        print('CONTINUING OLD PROCESS from build number: {}'.format(build_min))
        build_numb = int(build_min)
        
        if build_numb == 1:
            return
        

    while build_numb > 0:
        print('\t\tRequesting from build number: {}'.format(build_numb))

        endpoint = 'https://api.travis-ci.org/builds?after_number={}&repository_id={}'.format(build_numb,repo_id)
        r = requestTravisAPI(endpoint)
        
        if len(r) > 0:
            #print('Recording builds...')
            #print(r.json())
            insertBuilds(r,slug)
        
            build_numb = build_numb - 25
        else:
            break
```
