```python
def findAndSaveMergeConflicts(repo_name,pr_number,merge_commit_sha):
    merge_conflict = getMergeConflictOnCommits(repo_name,pr_number)
    
    #Is there merge conflict on Commits table?
    if len(merge_conflict) > 0:
        print('+ Merge conflict found.')
        updatePRMergeConflict(repo_name,pr_number)
        return
    
    #If there is no merge conflict, search for conflict on merge commit
    print('Searching for conflict on merge commit.')
    merge_conflict = getMergeConflictOnMerge(merge_commit_sha)
    if len(merge_conflict) > 0:
        print('+ Merge conflict found.')
        updatePRMergeConflict(repo_name,pr_number)
        return
    
    #If there is no merge conflict, search for conflict on pullrequest messages
    print('Searching for conflict on messages.')
    merge_conflict = getMergeConflictOnMessages(repo_name,pr_number)
    if len(merge_conflict) > 0:
        print('+ Merge conflict found.')
        updatePRMergeConflict(repo_name,pr_number)
```

```python
def loadTokens():
    f = open('/home/gh_tokens.txt', "r")
    tokens =[]
    tk = f.readline().replace('\n','')
    while tk != '':
        tokens.append(tk)
        tk = f.readline().replace('\n','')

    f.close()
    return tokens
```

```python
tokens = loadTokens()
i_token =0


def requestAPI(URL):
    # api-endpoint 
    global i_token
    
    if (len(tokens)-1) > i_token:
        i_token += 1
    else:
        i_token = 0
        
    r = requests.get(url = URL, headers={'Authorization': tokens[i_token],'Accept':'application/vnd.github.cloak-preview'}).json()
    if len(r) > 0:
        try:
            #print('keys: {}'.format(r.keys()))
            
            if isinstance(r, dict) and 'message' in r.keys():
                print('{} -- {}'.format(datetime.now().strftime("%H:%M:%S"),r['message']))
                if 'API rate limit exceeded' in r['message']:
                    time.sleep(600)
                    requestAPI(URL)
                else:
                    return None

            return r
        except Exception as e:
            print('\n Erro no request get: {}'.format(e))
            print(r)
    else:
        return None
```
