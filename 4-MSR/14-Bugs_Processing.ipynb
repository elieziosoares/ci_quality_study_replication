# RQ2 Projects Selection

- Get all projects from minor group (CI x Non-CI)
- Draw the same quantity from the other group

```python
qty_projects = getQtyProjects()

if qty_projects[0][1] < qty_projects[1][1]:
    setSelectedAllProjects(qty_projects[0][0])
    drawProjects(qty_projects[1][0],qty_projects[0][1])
else:
    setSelectedAllProjects(qty_projects[1][0])
    drawProjects(qty_projects[0][0],qty_projects[1][1])
```

```python
def drawProjects(ci,qty):
    projects = getGroupProjects(ci)
    projects = list(map(lambda x: x[0], projects))
    i=0
    
    while i < qty:
        random.shuffle(projects)
        id = random.randint(0, len(projects)-1)
        proj = projects.pop(id)
        
        i +=1
        
        print('PROJECT: {}'.format(proj))
        try:
            query = """UPDATE  rq2
                set rq2_included = true
                WHERE repo_name like %s"""

            connection = connectDB()
            cursor = connection.cursor()
            cursor.execute(query,[proj])
            connection.commit()
            cursor.close()
            connection.close()
        except psycopg2.IntegrityError as e:
            print ("==============================================================")
            print ("Error while updating into PostgreSQL. drawProjects >>> Exception: {}".format(e)) 
            connection.close()
        except Exception as e:
            print ("==============================================================")
            print ("Error while processing drawProjects >>> Exception: {}".format(e)) 
```

```python
def setSelectedAllProjects(ci):
    try:
        query = """UPDATE  rq2
            set rq2_included = true
            WHERE ci is %s"""

        connection = connectDB()
        cursor = connection.cursor()
        cursor.execute(query,[ci])
        connection.commit()
        cursor.close()
        connection.close()
    except psycopg2.IntegrityError as e:
        print ("==============================================================")
        print ("Error while updating into PostgreSQL. setSelectedAllProjects >>> Exception: {}".format(e)) 
        connection.close()
    except Exception as e:
        print ("==============================================================")
        print ("Error while processing setSelectedAllProjects >>> Exception: {}".format(e)) 
```

```python
def getGroupProjects(ci):
    query = """SELECT repo_name From rq2 WHERE ci is %s;"""

    
    connection = connectDB()
    cursor = connection.cursor()
    cursor.execute(query,[ci])
    result = cursor.fetchall()
    connection.close()
    return result
```

```python
def getQtyProjects():
    query = """select ci,count(repo_name) from RQ2 group by ci;"""

    connection = connectDB()
    cursor = connection.cursor()
    cursor.execute(query)
    result = cursor.fetchall()
    connection.close()
    return result
```

```python
def getProjects():
    query = """select P.repo_name, min(M.init_period), max(M.end_period), ci 
    FROM projects P INNER JOIN metrics_period M ON P.repo_name = M.repo_name
    WHERE P.rq1_included is true 
    group by P.repo_name ORDER BY repo_name;"""
    
    connection = connectDB()
    cursor = connection.cursor()
    cursor.execute(query)
    result = cursor.fetchall()
    connection.close()
    return result
```

```python
def getBugs(project,init,end):
    query = """SELECT issue_number,created_at,body FROM ISSUE 
    where repo_name like %s and isbug is true and created_at between %s and %s
    order by created_at;"""
    
    connection = connectDB()
    cursor = connection.cursor()
    cursor.execute(query,[project,init,end])
    result = cursor.fetchall()
    connection.close()
    return result
```

```python
def getBugsRQ2(project,init,end):
    connection = connectDB()
    query = """SELECT period,days_after_tag,days_before_tag,days_between_tags,size,haslink,hasimage,comments FROM ISSUE 
    where repo_name like %s and isbug is true 
    and size is not null and days_between_tags > 0
    and created_at between %s and %s
    order by created_at;"""
    
    df = pd.read_sql_query(query,con=connection,params=[project,init,end])
    connection.close()
    
    return df
```

```python
def getPreviousTag(project,bug_creation):
    query = """select node_id,object_date from tag 
    where repo_name like %s and object_date < %s
    order by object_date desc limit 1;"""
    
    connection = connectDB()
    cursor = connection.cursor()
    cursor.execute(query,[project,bug_creation])
    result = cursor.fetchone()
    connection.close()
    return result
```

```python
def getNextTag(project,bug_creation):
    query = """select node_id,object_date from tag 
    where repo_name like %s and object_date > %s
    order by object_date asc limit 1;"""
    
    connection = connectDB()
    cursor = connection.cursor()
    cursor.execute(query,[project,bug_creation])
    result = cursor.fetchone()
    connection.close()
    return result
```

```python
def updateProject(project,bugs):
    try:
        query = """UPDATE  projects 
            set qty_bugs_period = %s
            WHERE repo_name like %s;
            """

        connection = connectDB()
        cursor = connection.cursor()
        cursor.execute(query,[bugs,project])
        connection.commit()
        cursor.close()
        connection.close()
    except psycopg2.IntegrityError as e:
        print ("==============================================================")
        print ("Error while updating into PostgreSQL. updateProject >>> Exception: {}".format(e)) 
        connection.close()
    except Exception as e:
        print ("==============================================================")
        print ("Error while processing updateProject >>> Exception: {}".format(e)) 
```

```python
def updateBug(project,number,p_tag,n_tag,days_after,days_before,period,days_between_tags):
    try:
        query = """UPDATE  issue set 
                last_tag = %s,
                next_tag = %s,
                days_after_tag = %s,
                days_before_tag = %s,
                period = %s,
                days_between_tags = %s
            WHERE repo_name like %s and issue_number = %s;
            """

        connection = connectDB()
        cursor = connection.cursor()
        cursor.execute(query,[p_tag,n_tag,days_after,days_before,period,days_between_tags,project,number])
        connection.commit()
        cursor.close()
        connection.close()
    except psycopg2.IntegrityError as e:
        print ("==============================================================")
        print ("Error while updating into PostgreSQL. updateBug >>> Exception: {}".format(e)) 
        connection.close()
    except Exception as e:
        print ("==============================================================")
        print ("Error while processing updateBug >>> Exception: {}".format(e)) 
```

```python
def updateBugFlags(project,number,haslink,hasimage,size):
    try:
        query = """UPDATE  issue set 
                haslink = %s,
                hasimage = %s,
                size = %s
            WHERE repo_name like %s and issue_number = %s;
            """

        connection = connectDB()
        cursor = connection.cursor()
        cursor.execute(query,[haslink,hasimage,size,project,number])
        connection.commit()
        cursor.close()
        connection.close()
    except psycopg2.IntegrityError as e:
        print ("==============================================================")
        print ("Error while updating into PostgreSQL. updateBug >>> Exception: {}".format(e)) 
        connection.close()
    except Exception as e:
        print ("==============================================================")
        print ("Error while processing updateBug >>> Exception: {}".format(e)) 
```

```python
def processTags(project,number,created):
    days_after = days_before = days_between_tags = 0
    tag_before = tag_next = None
    
    p_tag = getPreviousTag(project,created)
    n_tag = getNextTag(project,created)
    
    if p_tag is not None:
        days_after = (created - p_tag[1]).days
        tag_before = p_tag[0]
        
    if n_tag is not None:
        days_before = (n_tag[1] - created).days
        tag_next = n_tag[0]
        
    if  p_tag is not None and n_tag is not None:
        days_between_tags = (n_tag[1] - p_tag[1]).days
    
    #by default the period is 2, i.e., nor close to previous release/tag, neither close to next tag.
    #If the bug was reported in the first part of the interval, set period = 1
    if days_after <= (days_between_tags/3) and days_between_tags > 0:
        period = 1
    elif days_before <= (days_between_tags/3) and days_between_tags > 0:
        period = 3
    else: 
        period = 2
        
    updateBug(project,number,tag_before,tag_next,days_after,days_before,period,days_between_tags)
```

```python
def processBody(project,number,body):
    size = len(body)
    images = links = False
    
    if body.count('https://user-images.com') > 0 or body.count('https://cloud.githubusercontent.com') > 0:
        images = True
        
        #If the bug has img and link, but the quantity of simple links is greater than images, 
        #then this bug report has links.
        qty_img = body.count('https://user-images.com') + body.count('https://cloud.githubusercontent.com')
        qty_link = body.count('http:') + body.count('https:')
        
        if qty_link > qty_img:
            links = True
    elif body.count('http:') > 0 or body.count('https:') > 0:
        links = True
        
    updateBugFlags(project,number,links,images,size)
```
